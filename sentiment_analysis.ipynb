```python
import pandas as pd
import math

# 读取CSV文件
df = pd.read_csv('datax1.csv')

# 按tweet_id分组，并计算加权平均数
grouped = df.groupby('tweet_id')

map1 = {
    'positive' : 1,
    'negative' : -1
}

def weighted_mean(group):
    # 计算赞数之和
    like_sum = group['like_counts'].sum()

    ret_mapped = group['ret'].apply(lambda x: map1.get(x.strip()))
    # 计算加权平均数
    weighted_sum = (ret_mapped * group['ret_prob'].apply(lambda x: math.exp(x)).fillna(0) * group['like_counts']).sum()
    # 计算加权平均数并赋值给新的一列
    group['weighted_mean'] = weighted_sum / like_sum
    # 将加权平均数转换为Series对象，并返回字典
    return pd.Series(group['weighted_mean'].values[0], index=[group['tweet_id'].values[0]])

# 计算加权平均数并写入CSV文件中
result = grouped.apply(weighted_mean).to_frame(name='weighted_mean').reset_index().drop(columns=['level_1'])
result.to_csv('analysis_result.csv', index=False)
result.to_excel('analysis_result.xlsx', index=False)
```

```python
import pandas as pd
import math

# 读取CSV文件
df = pd.read_csv('analysis_data.csv')

# 按tweet_id分组，并计算加权平均数
grouped = df.groupby('tweet_id')

map1 = {
    'positive' : 1,
    'negative' : -1
}

df['like_counts_scaled'] = (df['like_counts'] - df['like_counts'].min()) / (df['like_counts'].max() - df['like_counts'].min())

def weighted_mean(group):
    # 计算赞数之和
    like_sum = group['like_counts_scaled'].sum()

    ret_mapped = group['ret'].apply(lambda x: map1.get(x.strip()))
    # 计算加权平均数
    weighted_sum = (ret_mapped * group['ret_prob'].apply(lambda x: math.exp(x)).fillna(0) * group['like_counts_scaled']).sum()
    # 计算加权平均数并赋值给新的一列
    group['weighted_mean'] = weighted_sum / like_sum
    # 将加权平均数转换为Series对象，并返回字典
    return pd.Series(group['weighted_mean'].values[0], index=[group['tweet_id'].values[0]])

# 计算加权平均数并写入CSV文件中
result = grouped.apply(weighted_mean).to_frame(name='weighted_mean').reset_index().drop(columns=['level_1'])
result.to_csv('analysis_result_scaled.csv', index=False)
result.to_excel('analysis_result_scaled.xlsx', index=False)
```
